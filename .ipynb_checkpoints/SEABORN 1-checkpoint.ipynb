```python
# DISTPLOT (DISTRIBUTION PLOT) -- GIVES THE DENSITY OF DATA PASSED

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd

sb.distplot([1,2,3,4,5,6])

plt.show()
```

```python
data = np.random.randn(200)

sb.distplot(data)

plt.show()
```

```python
# LINE PLOT

x = [10,5,15,20,7,6]
y = [100,120,150,95,210,240]

data_plot = pd.DataFrame({"Product_no":x , "Cost":y})
data_plot

sb.lineplot(x='Product_no' , y='Cost' , data = data_plot)

# lineplot(DATA FOR X , DATA FOR Y , FROM WHERE TO TAKE THOSE X , Y DATAS)

plt.show()
```

```python
X = [10,24,63,24,74]
Y = [1,6,27,9,4]

sb.lineplot(X,Y)
plt.show()
```

```python
# LMPLOT

df = sb.load_dataset("anscombe")
print(df)

sb.lmplot(x='x' , y='y' , data=df)
plt.show()
```

```python
X = [10,24,63,24,74]
Y = [1,6,27,9,4]

sb.lmplot(X,Y) #NEED TO PASS 3 ARGUMENTS ... HENCE CONVERT IT INTO A DATAFRAME AND THEN PASS
plt.show()
```

```python
X = [10,24,63,24,74]
Y = [1,6,27,9,4]

DF = pd.DataFrame({'xlabel':X , 'ylabel':Y})

sb.lmplot(x='xlabel',y='ylabel',data=DF) 
plt.show()
```

```python
# MARGINAL PLOT (TOP AND RIGHT BARS SHOW DENSITY OR HISTOGRAM DISTRIBUTION OF TWO VARIABLES)

df = sb.load_dataset('iris')
print(df)

# CUSTOMISE THE INSIDE PLOT - THE OPTIONS ARE : "scatter" | "reg" | "resid" | "kde" | "hex"

sb.jointplot(x=df["sepal_length"], y=df["sepal_width"] , kind='scatter')
```

```python
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"] , kind='hex')
```

```python
# STRIP PLOT (ONE VARIABLE IS CATAGORICAL)

tips = sb.load_dataset('tips') #CREATES A DATAFRAME
print(tips)

sb.stripplot(x='day' , y='total_bill' , data=tips)
```

```python
# SWARM PLOT (SAME AS STRIP PLOT BUT POINTS DONT OVERLAP)

tips = sb.load_dataset('tips')
print(tips)

sb.swarmplot(x='day' , y='total_bill' , data=tips)
```

```python
# BOX PLOT 

sb.boxplot(x='day',y='total_bill',data=tips, palette='rainbow')
```

```python
# VIOLIN PLOT

sb.violinplot(x='day',y='total_bill',data=tips, palette='rainbow')
```

```python
data = np.random.random((20,6)) + np.arange(6)/2

#SOME THEMES ARE : darkgrid , whitegrid , dark , white and ticks

sb.set_style('darkgrid')

sb.boxplot(data=data)

plt.title("darkgrid")

plt.show()
```
