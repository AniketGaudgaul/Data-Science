```python
import pandas as pd
import numpy as np
```

```python
# CREATING A SERIES USING ARRAY

array = np.array(["a","b","c","d","e"])
series = pd.Series(array , index = [1,2,3,4,5])

print(series)
```

```python
# CREATING AN DATA FRAME USING LIST

list = [['aniket',60], ['john',67],['eric',83]]

DF = pd.DataFrame(list,columns=['Name','weight'], index=[1,2,3]) 

print(DF)
```

```python
# DATA FRAME USING A DICTIONARY

dict = {'NAME':['john' ,'eric','jason','jake'] , 'AGE' : [23,42,12,43]}

DF2 = pd.DataFrame(dict , index = [1,2,3,4])

print(DF2)
```

```python
dict2 = [{'a':2 , 'b':3},{'a':4 , 'b':5 , 'c':7}]

DF3 = pd.DataFrame(dict2)

print(DF3)
```

```python
# OPENING A CSV FILE

DF4 = pd.read_csv('C:/Users/LENOVO/Downloads/biostats.csv')

print(DF4)
```

```python
# READING A COLUMN OF CSV

print(DF4['Weight'])

#READING 2 COLUMNS

print(DF4[['Weight', 'Name']])
```

```python
# READING A COLUMN IN A DIFFERENT WAY WITH NAME OF THE COLUMN

DF4[['Name' , 'Weight']]
```

```python
# SUMMATION OF ALL DATA UNDER A COLUMN

sum(DF4["Weight"])
```

```python
# SHOWING ONE SAMPLE ROW FROM THE FRAME

print(DF4.sample(n = 1))
```

```python
# ADDING A NEW COLUMN

Salary = [12,31,4,2,2,1,34,4,2,3,12,4,5,64,3,2,5,3]

DF4['Salary (laks)'] = Salary

DF4
```

```python
# DELETING A COLUMN

del DF4['Weight']

DF4
```

```python
# SORTING VALUES IN A COLUMN IN ASSCENDING ORDER

DF4.sort_values("Salary (laks)") #NOTE OTHER COLUMNS GET ADJUSTED ACCORDINGLY
```

```python
# SORTING VALUES IN A COLUMN IN DECENDING ORDER

DF4.sort_values("Salary (laks)" , ascending = False) 
```

```python
# COUNTING UNIQUE VALUES INSIDE A COLUMN

DF4.nunique()  # UNIQUE VALUE MEANS HOW MANY DIFFERENT NOS ARE THERE IN THAT COLUMN
```

```python
# USING LOC  ----- LOC IS USED FOR PERSONALISED LABELS WHEREAS ILOC FOR PROPER INDEXES STARTING WITH 0 

DF4.loc[1:5]
```

```python
DF4.loc[1:5 ,:]  #2nd parameter is for column
```

```python
# SUMMARY 

DF4.describe()
```

```python
# SHOWS TRUE AND FALSE FOR THE APPLIED CONDITION

DF4['Salary (laks)'] >= 10
```

```python
# FORMS A DATAFRAME FOR ONLY THE ROWS WHICH SATISFY THE CONDITION

DF4[DF4['Salary (laks)'] > 10]
```

```python
# SELECTING ROWS WITH ORIGINAL INDEXES ---- ILOC FOR EXACT COLUMN AND ROW

DF4.iloc[[0,1,2,3,4,5]]
```

```python
DF4.iloc[0:5]
```

```python
# SELECTING ROWS WITH PARTICULAR INDICES AND PARTICULAR COLUMN

DF4.iloc[[0,2,5,4],[2,4]]
```
