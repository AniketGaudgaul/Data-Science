```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# HISTOGRAM -- SHOWS FREQUENCY OF A RANGE(BIN)

arr = np.array([22,55,24,74,24,53,78,99,77,96,56,34,53,63,72])

# plt.xlabel = ('Range of marks')
# plt.ylabel = ('No of students / frequency')

plt.hist(arr , bins = [0,25,50,75,100]) #it means 0-24 , 25-49 ....

plt.show()

#IF THE NUMBER IS LIKE 25 THEN IT WILL BE COUNTED IN 25-50 RANGE
```

```python
# NOT NECESSARY TO SPECIFY BINS RANGE 

numbers = np.random.randint(1,100,100)
print(numbers)

plt.hist(numbers , bins = 10)

plt.show()
```

```python
# BOXPLOT

x = [100,120,140,160,180]

plt.boxplot(x)

plt.show()

# EXTREME LINES - LOWEST AND HIGHEST VALUES
# MEDIAN IS THE LINE DIVIDING THE BOX
# BOX ENDINGS SHOW UPPER AND LOWER QUARTILE
```

```python
# CONTOUR PLOT USED TO SHOW A 3D IN A 2D 

# Z = f(x,y)

# GENERATING 100 NUMBERS FROM -3.0 TO 3.0

xlist = np.linspace(-3.0 , 3.0 , 100)
ylist = np.linspace(-3.0 , 3.0 , 100)

# CREATES A RECTANGULAR GRID OF X AND Y

X , Y = np.meshgrid(xlist, ylist)

Z = np.sqrt(X**2 + Y**2)

cp = plt.contourf(X , Y , Z)

# ADDS COLOR

plt.colorbar(cp)

plt.show()
```
