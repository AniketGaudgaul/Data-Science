# CAR PRICE PREDICTION MODEL

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bmw.csv')
data.head()
```

```python
data.info()
```

```python
null = data.isnull().sum()
null
```

No null values

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns = ['model','transmission','fuelType']

for i in columns:
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data.describe()
```

```python
X = data.iloc[:,0:8]
Y = data.iloc[:,8]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.20,random_state = 0)
```

```python
X_train
```

```python
%matplotlib inline

plt.figure(figsize=(17,7))

plt.subplot(2,4,1)
plt.scatter(X_train.iloc[:,0],Y_train)
plt.xlabel('model')
plt.ylabel('price')

plt.subplot(2,4,2)
plt.scatter(X_train.iloc[:,1],Y_train)
plt.xlabel('year')
plt.ylabel('price')

plt.subplot(2,4,3)
plt.scatter(X_train.iloc[:,2],Y_train)
plt.xlabel('transmission')
plt.ylabel('price')

plt.subplot(2,4,4)
plt.scatter(X_train.iloc[:,3],Y_train)
plt.xlabel('mileage')
plt.ylabel('price')

plt.subplot(2,4,5)
plt.scatter(X_train.iloc[:,4],Y_train)
plt.xlabel('fuelType')
plt.ylabel('price')

plt.subplot(2,4,6)
plt.scatter(X_train.iloc[:,5],Y_train)
plt.xlabel('tax')
plt.ylabel('price')

plt.subplot(2,4,7)
plt.scatter(X_train.iloc[:,6],Y_train)
plt.xlabel('mpg')
plt.ylabel('price')

plt.subplot(2,4,8)
plt.scatter(X_train.iloc[:,7],Y_train)
plt.xlabel('engineSize')
plt.ylabel('price')

plt.show()
```

# Co relations

# Creating a Pipeline if required add imputer to clear out missing values

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('std_scaler',StandardScaler()),
])
```

```python
import joblib
from joblib import dump,load

dump(my_pipeline,'my_pipeline')
```

```python
pipeline = joblib.load('my_pipeline')
```

```python
info = np.array([[4,2013,1,35000,0,240,43.5,3.0],[2,13,1,500,0,140,43.5,3.0]])

info_scaled = my_pipeline.fit_transform(info)

print(info_scaled)
```

```python
X_train_scaled = my_pipeline.fit_transform(X_train)
X_train_scaled
```

# Creating a model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train_scaled,Y_train)
```

```python
some_data = X_train.iloc[:5]
```

```python
some_labels = Y_train[:5]
prepared_data = my_pipeline.transform(some_data)
```

```python
prepared_data
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

```python
Y_train = np.array(Y_train)
Y_train
```

# Evaluating

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_train_scaled)

mse = mean_squared_error(Y_train,predictions)

rmse = np.sqrt(mse)
```

```python
rmse
```

```python
predictions
```

# Test data predictions

```python
X_test_scaled = my_pipeline.fit_transform(X_test)
model.fit(X_test_scaled,Y_test)
```

```python
predictions2 = model.predict(X_test_scaled)

mse = mean_squared_error(Y_test,predictions2)

rmse = np.sqrt(mse)

rmse
```

Thus our model has an root mean square error of around 1.5k which is quite less
